```python
#加载必要的库
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets,transforms
```

```python
#定义超参数
BATCH_SIZE= 128 #每批处理的数据可以自己定义，也可以是64或16，根据电脑性能决定
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
EPOCHS = 10 #训练数据集的轮次，可更改
```

```python
#构建pipeline，对图像做处理
pipeline = transforms.Compose([
    transforms.ToTensor(),#将图片转换成tensor张量
    transforms.Normalize((0.1307),(0.3081))#正则化，当模型训练过拟合时，可以降低模型的复杂度
])
```

```python
#下载，加载数据
from torch.utils.data import DataLoader
train_set = datasets.MNIST("data", train=True, download=True, transform=pipeline)
test_set = datasets.MNIST("data", train=False, download=True, transform=pipeline)

train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_set, batch_size=BATCH_SIZE, shuffle=True)
```

```python
#插入：如何读取data中的图片，直接打开是不行的，因为文件格式不对：原始二进制格式
with open("./data/MNIST/raw/train-images-idx3-ubyte","rb") as f:        #rb:Read Binary打开文件夹
    file = f.read()    #将整个文件的所有字节 (Bytes) 一次性读入内存，存放在变量 file 中
```

```python
#显示图片
image1 = [int(str(item).encode("ascii"),10) for item in file[16:16+784]]    #item：代表每一个像素的灰度值（0-255）
print(image1)
```

```python
import cv2    #OpenCV 主要承担 数据预处理 (Data Preprocessing) 和 图像增强 (Image Augmentation) 的任务 读写图像，几何变换，颜色处理，绘制标注
import numpy as np

image1_np = np.array(image1, dtype=np.uint8).reshape(28,28,1)   #这里的1是channel通道用于记录一种颜色的强度，由于是灰度图，故只需要一个，如果是彩色图就要用三个（三原色）

print(image1_np.shape)
```

```python
#构建网络模型
class Digit_without_dropout(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,10,5) #1:灰度图片的通道 ， 10：输出通道  ，5：kernel卷积核
        self.conv2 = nn.Conv2d(10,20,3) #10：输入通道  20：输出通道  3：kernel
        self.fc1 = nn.Linear(20*10*10,1024)  #20*10*10:输入通道，  1024：输出通道
        self.fc2= nn.Linear(1024,10)  #1024:输入通道，  10：输出通道

    def forward(self,x):
        input_size = x.size(0) #batch_size
        x = self.conv1(x) #输入：batch*1*28*28，  输出：batch*10*24*24(28-5+1)
        x = F.relu(x)  #激活层,保持shape不变：batch*10*24*24(28-5+1)
        x = F.max_pool2d(x,2,2)  #池化层：对图片进行压缩（降采样）（用于筛选） 输入：batch*10*24*24(28-5+1)  输出：batch*10*12*12  

        x = self.conv2(x) #输入：batch*10*12*12  输出：batch*20*10*10(12-3+1)
        x = F.relu(x)  #激活层,保持shape不变：batch*20*10*10

        x = x.view(input_size,-1) #flattern拉平（变成一条线） -1：自动计算维度：20*10*10=2000

        x = self.fc1(x) #输入：batch*2000  输出：batch*500
        x = F.relu(x)  #激活层,保持shape不变：batch*500

        x = self.fc2(x)  #输入：batch*500  输出：batch*10

        output = F.log_softmax(x,dim=1)  #计算分类后每个数字的概率值  dim=1表示按行计算

        return output
```

```python
#构建网络模型
class Digit_with_dropout(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1,10,5) #1:灰度图片的通道 ， 10：输出通道  ，5：kernel卷积核
        self.conv2 = nn.Conv2d(10,20,3) #10：输入通道  20：输出通道  3：kernel
        self.fc1 = nn.Linear(20*10*10,1024)  #20*10*10:输入通道，  1024：输出通道
        self.dropout = nn.Dropout(0.5)  
        self.fc2= nn.Linear(1024,10)  #1024:输入通道，  10：输出通道

    def forward(self,x):
        input_size = x.size(0) #batch_size
        x = self.conv1(x) #输入：batch*1*28*28，  输出：batch*10*24*24(28-5+1)
        x = F.relu(x)  #激活层,保持shape不变：batch*10*24*24(28-5+1)
        x = F.max_pool2d(x,2,2)  #池化层：对图片进行压缩（降采样）（用于筛选） 输入：batch*10*24*24(28-5+1)  输出：batch*10*12*12  

        x = self.conv2(x) #输入：batch*10*12*12  输出：batch*20*10*10(12-3+1)
        x = F.relu(x)  #激活层,保持shape不变：batch*20*10*10

        x = x.view(input_size,-1) #flattern拉平（变成一条线） -1：自动计算维度：20*10*10=2000

        x = self.fc1(x) #输入：batch*2000  输出：batch*500
        x = F.relu(x)  #激活层,保持shape不变：batch*500
        
        x = self.dropout(x)

        x = self.fc2(x)  #输入：batch*500  输出：batch*10

        output = F.log_softmax(x,dim=1)  #计算分类后每个数字的概率值  dim=1表示按行计算

        return output
```

```python
#定义优化器
model_a = Digit_without_dropout().to(DEVICE) #创建一个模型部署到设备上(实例化)
optimizer_a = optim.Adam(model_a.parameters())

model_b = Digit_with_dropout().to(DEVICE) #创建一个模型部署到设备上(实例化)
optimizer_b = optim.Adam(model_b.parameters())
```

```python
#定义训练方法
def train_model(model,train_loader,optimizer,device,epoch):
    total_loss = 0
    #模型训练
    model.train()
    for batch_index,(data,target) in enumerate(train_loader):
        #部署到device上去
        data, target = data.to(device) ,target.to(device)
        #梯度初始化为0
        optimizer.zero_grad()
        #训练结果
        output = model(data)
        #计算loss
        loss = F.cross_entropy(output,target) #交叉熵损失函数
        #反向传播
        loss.backward()
        #参数优化
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss/ len(train_loader)
    return avg_loss
```

```python
#定义测试方法
def test_model(model,device,test_loader):
    test_loss = 0
    #模型验证
    model.eval()
    #正确率
    correct = 0.0
    #测试损失
    test_loss = 0.0
    with torch.no_grad():  #不计算梯度也不会反向传播更新参数，这是训练集里面该做的事
        for data,target in test_loader:
            #部署到device上
            data, target = data.to(device), target.to(device)
            #测试数据
            output = model(data)
            #计算测试损失
            test_loss += F.cross_entropy(output,target).item()
            #找到概率最大值的下标
            pred = output.argmax(dim=1, keepdim = True)  #pred = output.max(1, keepdim=True)[1]  
                                                         #pred = torch.max(output,dim=1)有多种写法
            #累计正确的值
            correct += pred.eq(target.view_as(pred)).sum().item()
            
        test_loss /= len(test_loader.dataset)    
        accuracy = 100.0 * correct / len(test_loader.dataset)
        return test_loss,accuracy
```

```python
print("Start GroupA(without dropout)")
history_a = {'loss': [], 'acc': []}
for epoch in range(1, EPOCHS+1):
    train_model(model_a,train_loader,optimizer_a,DEVICE,epoch) #训练模型
    loss,acc = test_model(model_a,DEVICE,test_loader) #测试模型，获取当前准确率
    history_a['loss'].append(loss)
    history_a['acc'].append(acc)  
    
print("Start GroupB(with dropout)")
history_b = {'loss': [], 'acc': []}
for epoch in range(1, EPOCHS+1):
    train_model(model_b,train_loader,optimizer_b,DEVICE,epoch) #训练模型
    loss,acc = test_model(model_b,DEVICE,test_loader) #测试模型，获取当前准确率
    history_b['loss'].append(loss)
    history_b['acc'].append(acc)  
   
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))

# 1. 准确率对比图
plt.subplot(1, 2, 1)
plt.plot(history_a['acc'], label='No Dropout (512)', color='blue', linestyle='--')
plt.plot(history_b['acc'], label='With Dropout (1024)', color='red')
plt.title('Test Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.grid(True)

# 2. 测试 Loss 对比图 (看是否过拟合)
plt.subplot(1, 2, 2)
plt.plot(history_a['loss'], label='No Dropout (512)', color='blue', linestyle='--')
plt.plot(history_b['loss'], label='With Dropout (1024)', color='red')
plt.title('Test Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.savefig('comparison_results.png') # 保存这张对比图
plt.show()
```
